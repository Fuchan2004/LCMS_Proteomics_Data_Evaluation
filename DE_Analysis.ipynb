# Differential Expression Analysis

## Data Preparation

```python
#%pip install pydeseq2
```

```python
from pydeseq2.dds import DeseqDataSet
from pydeseq2.ds import DeseqStats

import pandas as pd
```

```python
counts = pd.read_csv('/vortexfs1/home/fadime.stemmer/0124_rpom_proteolysis/data/proteomes/DSS-3/Counts_240216_rpom_ProMM.csv')
counts
```

```python
# Set the first column to be the index rather than the number of lines
counts = counts.set_index('Identified Proteins (2758)')
counts
```

```python
# Get rid of columns that only have 0 counts
counts = counts[counts.sum(axis =1)>0]
counts
```

```python
# Transpose the Identified Proteins to be the columns, and the samples to be the rows (necessary for DESeq2)
counts = counts.T
counts
```

```python
# Need to have a metadata and set sample column our index.
metadata = pd.DataFrame(zip(counts.index,['ML','ML','ML','S','C']), columns=['Sample','Condition'])
metadata = metadata.set_index('Sample')
metadata
```

## Differential Expression Analysis

```python
dds = DeseqDataSet(counts=counts,
            colData=metadata,
            design_factors="Condition")
```

```python
dds.obs #observation dataframe = metadata
dds.X # countmatrix
dds.var # geneIDs
```

```python
dds.deseq2()
```
