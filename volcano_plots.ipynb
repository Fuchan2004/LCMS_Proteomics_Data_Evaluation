### P-values
Now onto the pvalue portion of our journey. We need to go through each row of the data and calculate the p value associated with our observations by using a t test. To do this, we'll use the ttest_ind function that we imported via the scipy.stats package. This function takes in two arrays (the individual values from each group) and returns a new array that contains the t statistic and the corresponding p value.

```python
# df1 and df2 are the dataframes containing the values of the triplicate measurements of treatment 1 and 2 respectively
def calculate_pvalue(df1, df2):
    num_values= df1.shape[0]
    p_values = []
    for row in range(num_values):
        ttest_result = stats.ttest_ind(df1.iloc[row].values, df2.iloc[row].values)
        p_values.append(ttest_result.pvalue)
    return p_values
```

```python
# Calculate the p-values and create a list for each treatment comparison
pval_DSS3_ML_AP = calculate_pvalue(DSS3_ACD_values, DSS3_ProMM_values)
pval_DSS3_S_AP = calculate_pvalue(DSS3stat_ACD_values, DSS3stat_ProMM_values)
pval_DSS3_ACD_MLS = calculate_pvalue(DSS3_ACD_values, DSS3stat_ACD_values)
pval_DSS3_ProMM_MLS = calculate_pvalue(DSS3_ProMM_values, DSS3stat_ProMM_values)
pval_DSS3_sec_AP = calculate_pvalue(DSS3sec_ACD_values, DSS3sec_ProMM_values)

pval_MIT1002_ML_AP = calculate_pvalue(MIT1002_ACD_values, MIT1002_ProMM_values)
pval_MIT1002_S_AP = calculate_pvalue(MIT1002stat_ACD_values, MIT1002stat_ProMM_values)
pval_MIT1002_ACD_MLS = calculate_pvalue(MIT1002_ACD_values, MIT1002stat_ACD_values)
pval_MIT1002_ProMM_MLS = calculate_pvalue(MIT1002_ProMM_values, MIT1002stat_ProMM_values)
pval_MIT1002_sec_AP = calculate_pvalue(MIT1002sec_ACD_values, MIT1002sec_ProMM_values)
```

```python
# Create a DataFrame from the p-values list
DF_pval_DSS3_ML_AP = pd.DataFrame(pval_DSS3_ML_AP, columns=["p_value_DSS3_ML_AP"])
DF_pval_DSS3_S_AP = pd.DataFrame(pval_DSS3_S_AP, columns=["p_value_DSS3_S_AP"])
DF_pval_DSS3_ACD_MLS = pd.DataFrame(pval_DSS3_ACD_MLS, columns=["p_value_DSS3_ACD_MLS"])
DF_pval_DSS3_ProMM_MLS = pd.DataFrame(pval_DSS3_ProMM_MLS, columns=["p_value_DSS3_ProMM_MLS"])
DF_pval_DSS3_sec_AP = pd.DataFrame(pval_DSS3_sec_AP, columns=["p_value_DSS3_AP_sec"])

DF_pval_MIT1002_ML_AP = pd.DataFrame(pval_MIT1002_ML_AP, columns=["p_value_MIT1002_ML_AP"])
DF_pval_MIT1002_S_AP = pd.DataFrame(pval_MIT1002_S_AP, columns=["p_value_MIT1002_S_AP"])
DF_pval_MIT1002_ACD_MLS = pd.DataFrame(pval_MIT1002_ACD_MLS, columns=["p_value_MIT1002_ACD_MLS"])
DF_pval_MIT1002_ProMM_MLS = pd.DataFrame(pval_MIT1002_ProMM_MLS, columns=["p_value_MIT1002_ProMM_MLS"])
DF_pval_MIT1002_sec_AP = pd.DataFrame(pval_MIT1002_sec_AP, columns=["p_value_MIT1002_AP_sec"])
```

```python
# Function to calculate -log10 of p-values and add as new columns
def add_neg_log10_pvalues(df, pval_columns):
    for col in pval_columns:
        neg_log10_col_name = f"-log10_{col}"
        df[neg_log10_col_name] = -np.log10(df[col].replace(0, np.nan))
        df[neg_log10_col_name].fillna(0, inplace=True)  # Replace -inf with 0
    return df
```

```python
# Add -log10 transformed p-values to DSS3
completepval_DSS3_ML_AP = add_neg_log10_pvalues(DF_pval_DSS3_ML_AP, ["p_value_DSS3_ML_AP"])
completepval_DSS3_S_AP = add_neg_log10_pvalues(DF_pval_DSS3_S_AP, ["p_value_DSS3_S_AP"])
completepval_DSS3_ACD_MLS = add_neg_log10_pvalues(DF_pval_DSS3_ACD_MLS, ["p_value_DSS3_ACD_MLS"])
completepval_DSS3_ProMM_MLS = add_neg_log10_pvalues(DF_pval_DSS3_ProMM_MLS, ["p_value_DSS3_ProMM_MLS"])
completepval_DSS3_sec_AP = add_neg_log10_pvalues(DF_pval_DSS3_sec_AP, ["p_value_DSS3_AP_sec"])

# Add -log10 transformed p-values to MIT1002
completepval_MIT1002_ML_AP = add_neg_log10_pvalues(DF_pval_MIT1002_ML_AP, ["p_value_MIT1002_ML_AP"])
completepval_MIT1002_S_AP = add_neg_log10_pvalues(DF_pval_MIT1002_S_AP, ["p_value_MIT1002_S_AP"])
completepval_MIT1002_ACD_MLS = add_neg_log10_pvalues(DF_pval_MIT1002_ACD_MLS, ["p_value_MIT1002_ACD_MLS"])
completepval_MIT1002_ProMM_MLS = add_neg_log10_pvalues(DF_pval_MIT1002_ProMM_MLS, ["p_value_MIT1002_ProMM_MLS"])
completepval_MIT1002_sec_AP = add_neg_log10_pvalues(DF_pval_MIT1002_sec_AP, ["p_value_MIT1002_AP_sec"])

# Replace NaN with 0
completepval_DSS3_ML_AP.fillna(0, inplace=True)
completepval_DSS3_S_AP.fillna(0, inplace=True)
completepval_DSS3_ACD_MLS.fillna(0, inplace=True)
completepval_DSS3_ProMM_MLS.fillna(0, inplace=True)
completepval_DSS3_sec_AP.fillna(0, inplace=True)
completepval_MIT1002_ML_AP.fillna(0, inplace=True)
completepval_MIT1002_S_AP.fillna(0, inplace=True)
completepval_MIT1002_ACD_MLS.fillna(0, inplace=True)
completepval_MIT1002_ProMM_MLS.fillna(0, inplace=True)
completepval_MIT1002_sec_AP.fillna(0, inplace=True)

#Extract as variable
transformed_pval_DSS3_ML_AP = completepval_DSS3_ML_AP["-log10_p_value_DSS3_ML_AP"]
transformed_pval_DSS3_S_AP = completepval_DSS3_S_AP["-log10_p_value_DSS3_S_AP"]
transformed_pval_DSS3_ACD_MLS = completepval_DSS3_ACD_MLS["-log10_p_value_DSS3_ACD_MLS"]
transformed_pval_DSS3_ProMM_MLS = completepval_DSS3_ProMM_MLS["-log10_p_value_DSS3_ProMM_MLS"]
transformed_pval_DSS3_sec_AP = completepval_DSS3_sec_AP["-log10_p_value_DSS3_AP_sec"]

transformed_pval_MIT1002_ML_AP = completepval_MIT1002_ML_AP["-log10_p_value_MIT1002_ML_AP"]
transformed_pval_MIT1002_S_AP = completepval_MIT1002_S_AP["-log10_p_value_MIT1002_S_AP"]
transformed_pval_MIT1002_ACD_MLS = completepval_MIT1002_ACD_MLS["-log10_p_value_MIT1002_ACD_MLS"]
transformed_pval_MIT1002_ProMM_MLS = completepval_MIT1002_ProMM_MLS["-log10_p_value_MIT1002_ProMM_MLS"]
transformed_pval_MIT1002_sec_AP = completepval_MIT1002_sec_AP["-log10_p_value_MIT1002_AP_sec"]

#Print to check
#print(completepval_DSS3_sec_AP)
#print(l2FC_DSS3_Sec_AP)
```

```python
def volcano_plot(plot_title, x_axis_title, y_axis_title, point_radius, log2_fold_change, transformed_pvalues, annotations):
    fig = go.Figure()
    fig.update_layout(
        title=plot_title,
        xaxis_title=x_axis_title,
        yaxis_title=y_axis_title,
        paper_bgcolor='white',
        plot_bgcolor='white',
    )

    colors = []

    for i in range(len(log2_fold_change)):
        if transformed_pvalues[i] > 2:
            if log2_fold_change[i] > 0.5:
                colors.append('#db3232')
            elif log2_fold_change[i] < -0.5:
                colors.append('#3f65d4')
            else:
                colors.append('rgba(150,150,150,0.5)')
        else:
            colors.append('rgba(150,150,150,0.5)')

    fig.add_trace(
        go.Scattergl(
            x=log2_fold_change,
            y=transformed_pvalues,
            mode='markers',
            text=annotations,
            hovertemplate='%{text}: %{x}<br>',
            marker={
                'color': colors,
                'size': point_radius,
            }
        )
    )
    fig.show()
```
